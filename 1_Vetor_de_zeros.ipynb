```python
#Atividade 37

import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])


x_mean = np.mean(x)
y_mean = np.mean(y)


B1 = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean) ** 2)


B0 = y_mean - B1 * x_mean

print(f"Slope (B1): {B1}")
print(f"Intercept (B0): {B0}")


y_pred = B0 + B1 * x
print("Predicted values:", y_pred)
```

```python
#Atividade 37

import numpy as np

np.random.seed(0)
X = np.random.rand(100, 2)
K = 3
centroides = X[np.random.choice(X.shape[0], K, replace=False)]

def dist_euclidiana(a, b):
    return np.linalg.norm(a - b, axis=1)

for i in range(10):

    distancias = np.array([dist_euclidiana(X, c) for c in centroides])
    clusters = np.argmin(distancias, axis=0)


    novos_centroides = np.array([X[clusters == j].mean(axis=0) for j in range(K)])


    if np.allclose(centroides, novos_centroides):
        break

    centroides = novos_centroides

print("Centróides finais:", centroides)
```

```python
#Atividade 38
import numpy as np

np.random.seed(0)
X = np.random.rand(100, 2)
X_centered = X - np.mean(X, axis=0)

cov_matrix = np.cov(X_centered.T)

autovalores, autovetores = np.linalg.eig(cov_matrix)

projecao_1D = X_centered.dot(autovetores[:, 0])

X_reconstruido = projecao_1D[:, np.newaxis] * autovetores[:, 0]

print("Autovetores:", autovetores)
print("Dados projetados em 1D:", projecao_1D[:5])
```

```python
#Atividade 39
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view


imagem = np.random.rand(5, 5)
kernel = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]])

blocos = sliding_window_view(imagem, kernel.shape)

saida = np.tensordot(blocos, kernel, axes=((2, 3), (0, 1)))

print("Resultado da convolução:", saida)

```

```python
#Atividade 40
import numpy as np

A = np.random.rand(100, 100)
b = np.random.rand(100)

x = np.linalg.solve(A, b)

residuo = np.linalg.norm(np.dot(A, x) - b) / np.linalg.norm(b)

cond_A = np.linalg.cond(A)

print("Resíduo relativo:", residuo)
print("Condicionamento de A:", cond_A)
```

```python
#Atividade 41

import numpy as np

X = np.random.rand(100, 3)
y = np.random.rand(100)

coef, _, _, _ = np.linalg.lstsq(X, y, rcond=None)

y_pred = X.dot(coef)

rmse = np.sqrt(np.mean((y - y_pred) ** 2))

print("Coeficientes:", coef)
print("RMSE:", rmse)
```
